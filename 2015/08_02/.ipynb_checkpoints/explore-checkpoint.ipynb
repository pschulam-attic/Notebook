## Notes

3,250 unique patients

Many have non-unique "First Seen" dates. Same with first symptom dates.

```python
import pandas as pd
from dateutil.parser import parse
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams['figure.figsize'] = (8, 6)
```

```python
ptdata = pd.read_csv('ptdata.csv')
```

```python
ptid = ptdata['ingPtID']
```

```python
def strip_time(dates):
    for d in dates:
        try:
            yield d[:-15]
        except:
            yield d
            
first_seen = ptdata['strFrstSeen']
first_sick = ptdata['strDoFrstSx']

first_seen = pd.Series(strip_time(first_seen)).map(parse, 'ignore')
first_sick = pd.Series(strip_time(first_sick)).map(parse, 'ignore')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))
sns.distplot(first_seen.dt.year, ax=axes[0])
sns.distplot(first_seen.dt.month, ax=axes[1])
sns.distplot(first_seen.dt.day, ax=axes[2])
```

```python
from collections import Counter
counts = list(Counter(first_seen.dt.year).items())
year, number = zip(*counts)
```

```python
plt.scatter(year, number)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))
sns.distplot(first_sick.dt.year.dropna(), ax=axes[0])
sns.distplot(first_sick.dt.month.dropna(), ax=axes[1])
sns.distplot(first_sick.dt.day.dropna(), ax=axes[2])
```
